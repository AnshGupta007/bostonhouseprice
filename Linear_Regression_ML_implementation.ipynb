```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
boston_df = pd.read_csv('boston_housing.csv')
```

```python
boston_df.keys()
```

```python
dataset = pd.DataFrame(boston_df)
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
## exploratory data analysis
# correlation
dataset.corr()
```

```python
dataset.rename(columns={'medv': 'Price'}, inplace=True)
```

```python
plt.scatter(dataset['crim'], dataset['Price'])
```

```python
import seaborn as sns
sns.regplot(x="rm", y="Price", data=dataset)
```

```python
# diving our features into dependent and independent features
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## on which parameter the model has been trained
regression.get_params()
```

```python
## prediction with test data
reg_pred = regression.predict(X_test)
```

```python
reg_pred
```

```python
# plot a scatter plot for the prediction
plt.scatter(y_test, reg_pred)
```

```python
residuals = y_test - reg_pred
```

```python
residuals
```

```python
# plot these residuals
sns.displot(residuals, kind='kde')
```

```python
# scatter plot with respect to predictions and residuals uniform distribution
plt.scatter(reg_pred, residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test, reg_pred))
print(mean_squared_error(y_test, reg_pred))
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
# r square and adjusted r square
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```
